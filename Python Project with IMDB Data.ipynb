```python
#Import libraries that we're going to use.

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline 
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjusting the configurations of the plots we will create.


#Read in the data we downloaded from Kaggle. (https://www.kaggle.com/datasets/danielgrijalvas/movies)

df=pd.read_csv(r'C:\Users\Prophiii\Desktop\movies.csv')
```

```python
#Display of the data.

df
```

```python
#Using the loop function to see if there are any missing data.

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
#Removing the missing values from the data. Double checking if it's worked or not.

df = df.dropna()

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
#Data types of the columns.

df.dtypes
```

```python
#Getting rid of any duplicates.

df['company'].drop_duplicates()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#Visualization using Scatter plot with budget vs gross earnings of the movies.

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Budget for Film')

plt.ylabel('Gross Earnings')

plt.show()
```

```python
#Plot budget vs gross using Seaborn.

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'red'}, line_kws={'color':'black'})
```

```python
#Correlation matrix between all numeric columns.

df.corr(method='pearson')
```

```python
 df.corr(method='kendall')
```

```python
df.corr(method='spearman')
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```

### As can be seen, the budget of the movie and user votes on the platform has the highest correlation to gross earnings.
